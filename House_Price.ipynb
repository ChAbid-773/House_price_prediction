#**Seaborn Plot**

```
sns.scatterplot(x=hp['property_type'], y=hp['Area_in_Marla'])
plt.title('Scatter')
plt.xlabel('property_type')
plt.ylabel('Area_in_Marla')
plt.show()
```

```
sns.lineplot(x=hp['property_type'], y=hp['Area_in_Marla'])
plt.title('Line')
plt.xlabel('property_type')
plt.ylabel('Area_in_Marla')
plt.show()
```

```
sns.barplot(x=hp['property_type'], y=hp['Area_in_Marla'])
plt.title('Bar')
plt.xlabel('property_type')
plt.ylabel('Area_in_Marla')
plt.show()
```

```
sns.boxplot(x=hp['property_type'], y=hp['Area_in_Marla'])
plt.title('Box')
plt.xlabel('property_type')
plt.ylabel('Area_in_Marla')
plt.show()
```

```
sns.violinplot(x=hp['property_type'], y=hp['Area_in_Marla'])
plt.title('Violin')
plt.xlabel('property_type')
plt.ylabel('Area_in_Marla')
plt.show()
```

```
sns.kdeplot(x=hp['property_type'], y=hp['Area_in_Marla'])
plt.title('KDE')
plt.xlabel('property_type')
plt.ylabel('Area_in_Marla')
plt.show()
```

```
sns.pairplot(hp[['property_type', 'Area_in_Marla']])
plt.title('Pair')
plt.xlabel('property_type')
plt.ylabel('Area_in_Marla')
plt.show()
```
